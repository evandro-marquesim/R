```r
### modelo de regressao
### simulando os dados do proc. gerador
n = 100
set.seed(232)
x = seq(0,10,length=n)
y = rep(0,n)
b0 = 2
b1 = 1
dp = 1
e = rnorm(n,0,dp)
for(i in 1:n){
  y[i] = b0 + b1*x[i] + e[i]
}
z = b0 + b1*x + e
length(y)
z-y
par(mfrow=c(1,1))
plot(x,y)
d = y-z
mean(d)
### estimando a reg. para os dados gerados
m1 = lm(y~x)   ## linear model - objeto m1 guarda todos resultados
summary(m1)
m1$coef[2]*2
res = residuals(m1)
plot(x,res)
res2 = m1$residuals
mean(res-res2)
m1$coef[2]+10
abline(m1,col="blue")
```

# Comparação de estimadores


```r
set.seed(347)


n=100 #tamanho da amostra
nr=10000 #número de réplicas
teta=2 #parâmetro #duvida
tec1=rep(0,nr) # ̂1 repete 0 nr vezes
tec2=tec1


```

```r
for(i in 1:nr){
 x=runif(n,0,teta)
 tec1[i]=max(x) #Máxima Verossimilhança
 tec2[i]=2*mean(x) #Método dos Momentos
}
# duvida sobre o 2*mean(x)
#tec1 realizou maximizou a variável aleatória x proveniente de uma uniforme (n=100,0,2) da amostra 1 a nr=1000
#tec2 realizou o dobro da média de x da amostra 1 a nr=1000.
```

```r
e1=tec1-teta
e2=tec2-teta
eqm1=sum(e1^2)/nr #Erro Quadrático Médio de e1
eqm2=sum(e2^2)/nr #Erro Quadrático Médio de e2

boxplot(e1,e2)
```

```r
set.seed(347)

n=c(10,100,1000) #tamanho da amostra em vetor com 3 tamanhos diferentes (10,100,1000)
nr=10000 #número de réplicas
nn=length(n) #comprimento de n
teta=2 #Valor do parâmetro
tec1=matrix(rep(0,nn*nr),ncol=nn) # ̂1 repete 0 nr vezes
tec2=tec1
eqm1=rep(0,nn)
eqm2=eqm1
```

```r
for(i in 1:nn){
 for(j in 1:nr){
 x=runif(n[i],0,teta)
 tec1[j,i]=max(x) #Máxima Verossimilhança
 tec2[j,i]=2*mean(x) #Método dos momentos
}
}
e1=tec1-teta
e2=tec2-teta
for(i in 1:nn){
eqm1[i]=sum(e1[,i]^2)/nr
eqm2[i]=sum(e2[,i]^2)/nr
}
boxplot(e1[,1],e2[,1],e1[,2],e2[,2],e1[,3],e2[,3]) 

eqm1
eqm2
```

# bootstrap

```r
# X segue uma distribuição poisson(30) - mas n sabemos dessa informação
# número de tornados por ano em uma cidade qualquer 
amostra <- rpois(n = 200, lambda = 30)
amostra
```

```r
xbarra <- mean(amostra)
xbarra
```

```r
# standard error 
# The standard error of the mean, or simply standard error, 
# indicates how different the population mean is likely to be from a sample mean. 
# It tells you how much the sample mean would vary if you were to repeat a study using new samples from within a single population.

#sqrt(var(x))/sqrt(n)
#nesse caso, a variancia é igual a média pois é uma poisson
sqrt(30)/sqrt(200)
```

```r
# selecionando uma amostra para o bootstrap

sample(amostra, replace = T)

# bootstrap

bootstrap <- replicate(n = 1000, sample(amostra, replace = T))
```

```r
dim(bootstrap)
```

```r
medias <- apply(bootstrap, MARGIN = 2, FUN = mean)
medias
```

```r
#standard error do bootstrap

sqrt(sum((medias-xbarra)^2)/ncol(bootstrap))
```
